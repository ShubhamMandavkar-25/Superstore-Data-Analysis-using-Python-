## Feature Engineering

```python
df.head(3)
```

```python
df['Order_Year']=df['Order Date'].dt.year
```

```python
df['Order_Month']=df['Order Date'].dt.month
```

```python
df['Profit_Margin'] = df['Profit'] / df['Sales']
```

```python
df['Revenue_per_Quantity'] = df['Sales'] / df['Quantity']
```

```python
df['Order_Processing_Time']=(df['Ship Date']- df['Order Date']).dt.days
```

```python
cat_cols = ['Segment', 'Category', 'Region', 'Ship Mode']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
df['Discount_Percentage'] = df['Discount'] * 100
```

# Visualization

#### 1  Yearly Sales Trend

```python
yearly = df.groupby('Order_Year')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
plt.plot(yearly['Order_Year'], yearly['Sales'], marker='o', color='teal')
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.xticks(yearly['Order_Year']) 
plt.show()
```

#### 2  Monthly Sales Trend

```python
monthly = df.groupby('Order_Month')['Sales'].sum().reset_index()

import plotly.express as px

fig = px.line(monthly, x='Order_Month', y='Sales',title="Monthly Sales Trend", markers=True, labels={'Order_Month': 'Month', 'Sales': 'Total Sales'})
fig.update_layout( width=900, height=500, xaxis=dict(dtick=1) )

fig.show()
```

#### 3 Total Sales by Category

```python
sales_cat = df.groupby('Category')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
bars = plt.bar(sales_cat['Category'], sales_cat['Sales'], color='skyblue')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.bar_label(bars)

plt.show()
```

#### 4 Profit by Category

```python
profit_cat = df.groupby('Category')['Profit'].sum().reset_index()

plt.figure(figsize=(8,5))
pt = sns.barplot(data=profit_cat, x='Category', y='Profit', palette='viridis')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

for i in pt.containers:
    pt.bar_label(i, fmt='%.0f')

plt.show()
```

#### 5 Profit by Region

```python
profit_region = df.groupby('Region')['Profit'].sum().reset_index()

fig = px.bar(profit_region, x='Region', y='Profit', color='Region',
             title="Profit by Region", text_auto='.0f')
fig.show()
```

#### 6 Sales by Region

```python
sales_region = df.groupby('Region')['Sales'].sum().reset_index()

fig = px.pie(sales_region,names='Region',values='Sales',title="Sales by Region",width=800,height=600)
fig.update_traces(textposition='outside', textinfo='label+percent')

fig.show()
```

####  7 Sales by Customer Segment

```python
sales_segment = df.groupby('Segment')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sl=sns.barplot(data=sales_segment, x='Segment', y='Sales', palette='coolwarm')
plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")

for i in sl.containers:
    sl.bar_label(i, fmt='%.0f')
    
plt.show()
```

#### 8 Top 10 Products by Sales

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sls = sns.barplot(x=top_products.values, y=top_products.index, palette='Blues_r')
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")

for i in sls.containers:
    sls.bar_label(i, fmt='%.0f')

plt.show()
```

####  9 Bottom 10 Products by Profit

```python
bottom_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10).reset_index()
fig = px.bar(
    bottom_products,
    x='Product Name',
    y='Profit',
    color='Profit',
    color_continuous_scale='Reds_r',
    title="Top 10 Loss-Making Products",
    text_auto='.2s',
    width=1500,
    height=800
)

fig.update_layout(xaxis_tickangle=-45)

fig.show()
```

#### 10 Discount vs Profit Margin

```python
fig = px.scatter(df, x='Discount_Percentage', y='Profit_Margin',
                 title="Discount vs Profit Margin",
                 labels={'Discount_Percentage': 'Discount (%)', 'Profit_Margin': 'Profit Margin'},
                 color='Category')
fig.show()
```

#### 11 Order Processing Time by Ship Mode

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Ship Mode', y='Order_Processing_Time', palette='Set2')
plt.title("Order Processing Time by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Days")
plt.show()
```

# Summary

### Data Analysis Key Findings

*  **Yearly Sales Trend:** The line chart show yearly sales trend chart, total sales slightly decline from 2014 to 2015 but then rise sharply in 2016 and 2017, indicating strong recovery and accelerating growth over the period.
*  **Monthly Sales Trend:** This monthly sales chart for your dataset shows relatively low revenue in January and February, a sharp jump in March, steady mid-year performance from April to August, and very strong peaks from September through December, highlighting pronounced end-of-year growth and clear seasonality in customer demand.
*  **Total Sales by Category:** The sales by category bar chart shows that Technology records the highest sales, Furniture follows closely behind, and Office Supplies has slightly lower sales than the other two categories, indicating all three contribute strongly with Technology leading overall.
*  **Profit by Category:** The bar chart show that Technology generates the highest profit, Office Supplies delivers strong but slightly lower profit, and Furniture lags far behind, showing it is the least profitable category in the portfolio.
*  **Profit by Region:** The profit by region bar chart shows that the West is the most profitable region, followed by the East and South, while the Central region generates the lowest profit among all four regions.
*  **Sales by Region:** The sales by region pie chart shows that the West contributes the largest share of total sales, followed closely by the East, while the Central region provides a solid but smaller portion and the South accounts for the lowest share among all four regions.
*  **Sales by Customer Segment:** The “Sales by Customer Segment” bar chart shows that the Consumer segment contributes the highest sales, followed by Corporate, while Home Office generates the lowest revenue, indicating Consumers are the primary driver of overall sales performance.
*  **Top 10 Products by Sales:** The barchart shows top 10 products by sales, showing which items contribute the most revenue and should be prioritized for inventory focus, marketing, and strategic decision-making.
*  **Bottom 10 Products by Profit:** This chart highlights the bottom 10 products by profit, showing that a small group of mainly high-cost office equipment and furniture items are generating the largest losses and should be prioritized for review or corrective action.
*  **Discount vs Profit Margin:** The scatter plot shows that higher discounts, especially on Furniture and Office Supplies, are associated with lower or negative profit margins
*  **Order Processing Time by Shipmode:**  The box plot shows that faster ship modes like Same Day and First Class have much shorter and less variable order processing times than Second Class and Standard Class.
